```python
import math

class PhysicsSolver:
    def __init__(self):
        # Universal Physical Constants
        self.k = 8.987551e9        # Coulomb's constant (N m^2 / C^2)
        self.eps_0 = 8.854187e-12  # Vacuum permittivity (C^2 / N m^2)
        self.mu_0 = 4 * math.pi * 1e-7 # Vacuum permeability (T m / A)
        self.e_charge = 1.602e-19  # Elementary charge (C)
        self.m_e = 9.109e-31       # Electron mass (kg)
        self.c = 2.9979e8          # Speed of light (m/s)

    def print_header(self, text):
        print(f"\n{'='*40}\n{text}\n{'='*40}")

    def solve_week_1(self):
        self.print_header("Week 1: Electric Fields")
        
        # Problem 1
        q = 3.0e-9; r = 0.20
        E = self.k * q / (r**2)
        print(f"1) Electric Field E = {E:.2f} N/C")

        # Problem 2
        q1 = 4.0e-9; x1 = 0.0
        q2 = -2.0e-9; x2 = 0.30
        x_target = 0.10
        E1 = self.k * q1 / ((x_target - x1)**2) # Points +x (away from positive)
        E2 = self.k * abs(q2) / ((x2 - x_target)**2) # Points +x (towards negative)
        E_tot = E1 + E2
        print(f"2) E_net = {E_tot:.2f} N/C in the +x direction")

        # Problem 3
        d = 1.5e-2; delta_V = 450
        E_uniform = delta_V / d
        # Work-Energy Theorem: W = q*V = 0.5*m*v^2
        v = math.sqrt(2 * self.e_charge * delta_V / self.m_e)
        print(f"3a) Uniform Field E = {E_uniform:.2f} V/m")
        print(f"3b) Electron speed v = {v:.2e} m/s")

    def solve_week_2(self):
        self.print_header("Week 2: Gauss's Law")
        
        # Problem 1
        q = 8.0e-9
        flux = q / self.eps_0
        print(f"1) Electric Flux = {flux:.2f} N*m^2/C")

        # Problem 2
        R = 0.05; Q = 2.5e-6
        r_in = 0.02; r_out = 0.08
        E_in = (self.k * Q * r_in) / (R**3)
        E_out = (self.k * Q) / (r_out**2)
        print(f"2a) E inside (r=2cm) = {E_in:.2e} N/C")
        print(f"2b) E outside (r=8cm) = {E_out:.2e} N/C")

        # Problem 3
        sigma = 6.0e-9
        E_sheet = sigma / (2 * self.eps_0)
        print(f"3) E between sheet and plate = {E_sheet:.2f} N/C")
        print(f"   E beyond conducting plate = {E_sheet:.2f} N/C (assuming uncharged plate polarizes)")

    def solve_week_3(self):
        self.print_header("Week 3: Electric Potential")
        
        # Problem 1
        q = 5.0e-9; r = 0.30
        V = self.k * q / r
        print(f"1) Potential V = {V:.2f} V")

        # Problem 2
        q1 = 3.0e-9; q2 = -3.0e-9; r_mid = 0.20; q0 = 1.0e-9
        V_mid = (self.k * q1 / r_mid) + (self.k * q2 / r_mid)
        W = q0 * V_mid
        print(f"2) V at midpoint = {V_mid:.2f} V")
        print(f"   Work to bring q0 from infinity = {W:.2f} J")

        # Problem 3
        a = 12
        def E_field(x): return -2 * a * x
        q_move = 2.0e-9; x_i = 0.10; x_f = 0.40
        delta_V = (a * x_f**2) - (a * x_i**2)
        Work_external = q_move * delta_V # Work done BY external agent
        print(f"3a) E(x) = -dV/dx = -{2*a}x V/m")
        print(f"3b) Work required = {Work_external:.2e} J")

    def solve_week_4(self):
        self.print_header("Week 4: Capacitance and Dielectrics")
        
        # Problem 1
        A = 80e-4; d = 1.0e-3
        C = self.eps_0 * A / d
        print(f"1) Capacitance C = {C:.2e} F")

        # Problem 2
        C_uF = 6.0e-6; V = 12
        Q = C_uF * V
        U = 0.5 * C_uF * (V**2)
        # Disconnected -> Q is constant, C halves (since d doubles)
        C_new = C_uF / 2
        V_new = Q / C_new
        U_new = 0.5 * (Q**2) / C_new
        print(f"2a) Q = {Q:.2e} C")
        print(f"2b) U = {U:.2e} J")
        print(f"2c) New V = {V_new:.2f} V, New U = {U_new:.2e} J")

        # Problem 3
        C0 = 150e-12; kappa = 4.2
        # a) Connected to source
        V_conn = 200
        C_conn = kappa * C0
        Q_conn = C_conn * V_conn
        U_conn = 0.5 * C_conn * (V_conn**2)
        print(f"3a) Connected: C = {C_conn:.2e} F, Q = {Q_conn:.2e} C, U = {U_conn:.2e} J")
        # b) Isolated
        Q_iso = 30e-9
        C_iso = kappa * C0
        V_iso = Q_iso / C_iso
        U_iso = 0.5 * (Q_iso**2) / C_iso
        print(f"3b) Isolated: V = {V_iso:.2f} V, U = {U_iso:.2e} J")

    def solve_week_5(self):
        self.print_header("Week 5: Current and Resistance")
        
        # Problem 1
        R = 18; V = 9.0
        I = V / R
        P = V * I
        print(f"1) I = {I:.2f} A, P = {P:.2f} W")

        # Problem 2
        # Assuming electron density of copper n = 8.49e28 m^-3
        n = 8.49e28; A = 3.31e-6; I_drift = 10.0
        v_d = I_drift / (n * self.e_charge * A)
        print(f"2) Drift velocity v_d = {v_d:.2e} m/s")

        # Problem 3
        R0 = 5.0; alpha = 0.0060; delta_T = 120 - 20; V_hot = 24
        R_hot = R0 * (1 + alpha * delta_T)
        I_hot = V_hot / R_hot
        P_hot = (V_hot**2) / R_hot
        print(f"3) Heated: R = {R_hot:.2f} Ohms, I = {I_hot:.2f} A, P = {P_hot:.2f} W")

    def solve_week_6(self):
        self.print_header("Week 6: DC Circuits")
        
        # Problem 1
        R1 = 4; R2 = 6; V_series = 20
        Req_s = R1 + R2
        I_s = V_series / Req_s
        print(f"1) Series: Req = {Req_s} Ohms, I = {I_s:.2f} A")

        # Problem 2
        R1 = 3; R2 = 6; R3 = 9; V_p = 18
        Req_p = 1 / (1/R1 + 1/R2 + 1/R3)
        print(f"2) Parallel: Req = {Req_p:.2f} Ohms")
        print(f"   I_R1 = {V_p/R1:.2f} A, I_R2 = {V_p/R2:.2f} A, I_R3 = {V_p/R3:.2f} A")

        # Problem 3
        # Using typical 2-loop standard (Loop 1 left CW, Loop 2 right CW)
        # Eq 1: 12 - 4*I1 - 5*(I1 - I2) = 0  =>  9*I1 - 5*I2 = 12
        # Eq 2: -9 - 6*I2 - 5*(I2 - I1) = 0  => -5*I1 + 11*I2 = -9
        # Solving using Cramer's rule
        det = (9 * 11) - (-5 * -5)
        I1 = ((12 * 11) - (-9 * -5)) / det
        I2 = ((9 * -9) - (-5 * 12)) / det
        print(f"3) Kirchhoff: I1 (Left loop) = {I1:.3f} A, I2 (Right loop) = {I2:.3f} A")

    def solve_week_8(self):
        self.print_header("Week 8: Magnetic Fields")
        
        # Problem 1
        v = 3.0e6; B = 0.20
        F_mag = self.e_charge * v * B
        r_orbit = (self.m_e * v) / (self.e_charge * B)
        print(f"1) Force F = {F_mag:.2e} N, Radius r = {r_orbit:.2e} m")

        # Problem 2
        L = 0.40; I = 12; B_wire = 0.60; theta = math.radians(30)
        F_wire = I * L * B_wire * math.sin(theta)
        print(f"2) Force on wire = {F_wire:.2f} N")

        # Problem 3
        E = 2.5e4; B_sel = 0.80
        v_sel = E / B_sel
        print(f"3) Velocity for no deflection = {v_sel:.2e} m/s")

    def solve_week_9(self):
        self.print_header("Week 9: Sources of Magnetic Field")
        
        # Problem 1
        I = 15; r = 0.05
        B_straight = (self.mu_0 * I) / (2 * math.pi * r)
        print(f"1) B from straight wire = {B_straight:.2e} T")

        # Problem 2
        R = 0.08; I_loop = 6.0
        B_center = (self.mu_0 * I_loop) / (2 * R)
        print(f"2) B at loop center = {B_center:.2e} T")

        # Problem 3
        n = 1200; I_sol = 1.8
        B_sol = self.mu_0 * n * I_sol
        u_B = (B_sol**2) / (2 * self.mu_0)
        print(f"3) Solenoid: B = {B_sol:.2e} T, Energy density u = {u_B:.2f} J/m^3")

    def solve_week_10(self):
        self.print_header("Week 10: Faraday's Law")
        
        # Problem 1
        N = 200; dFlux = 0.010 - 0.040; dt = 0.050
        emf = -N * (dFlux / dt)
        print(f"1) Induced EMF = {emf:.2f} V")

        # Problem 2
        L = 0.35; v = 4.0; B = 0.75; R = 2.0
        motional_emf = B * L * v
        I = motional_emf / R
        P = (motional_emf**2) / R
        print(f"2) Motional EMF = {motional_emf:.2f} V, Current = {I:.2f} A, Power = {P:.2f} W")

        # Problem 3
        L_ind = 25e-3; dI = 0 - 3.0; dt_ind = 8.0e-3
        emf_self = -L_ind * (dI / dt_ind)
        Energy_init = 0.5 * L_ind * (3.0**2)
        print(f"3) Induced EMF = {emf_self:.2f} V, Initial stored energy = {Energy_init:.3f} J")

    def solve_week_11(self):
        self.print_header("Week 11: Inductance")
        
        # Problem 1
        length = 0.20; A = 1.2e-4; N = 800
        L = (self.mu_0 * (N**2) * A) / length
        print(f"1) Solenoid Inductance L = {L:.2e} H")

        # Problem 2
        R = 12; L_RL = 0.48; eps = 24; t = 0.060
        tau = L_RL / R
        I_t = (eps / R) * (1 - math.exp(-t / tau))
        print(f"2) Time constant tau = {tau:.3f} s, I(0.06s) = {I_t:.2f} A")

        # Problem 3
        # Assuming series configuration as they undergo the same current increase
        L_eq = 30e-3 + 50e-3
        U_initial = 0.5 * L_eq * (2.0**2)
        U_final = 0.5 * L_eq * (5.0**2)
        delta_U = U_final - U_initial
        print(f"3) Change in energy = {delta_U:.2f} J")

    def solve_week_12(self):
        self.print_header("Week 12: AC Circuits")
        
        # Problem 1
        R = 120; V_rms = 220
        I_rms = V_rms / R
        P_avg = (I_rms**2) * R
        print(f"1) Pure Resistor: I_rms = {I_rms:.2f} A, P_avg = {P_avg:.2f} W")

        # Problem 2
        R = 200; C = 8.0e-6; f = 50; V_rms = 120
        omega = 2 * math.pi * f
        Xc = 1 / (omega * C)
        Z = math.sqrt(R**2 + Xc**2)
        I_rms = V_rms / Z
        phase = math.degrees(math.atan(-Xc / R))
        print(f"2) RC Circuit: Xc = {Xc:.2f} Ohms, Z = {Z:.2f} Ohms")
        print(f"   I_rms = {I_rms:.3f} A, Phase angle = {phase:.2f} degrees")

        # Problem 3
        R = 25; L = 0.35; C = 18e-6
        f0 = 1 / (2 * math.pi * math.sqrt(L * C))
        Q_factor = (1 / R) * math.sqrt(L / C)
        print(f"3) RLC Resonance: f0 = {f0:.2f} Hz, Q-factor = {Q_factor:.2f}")

    def solve_week_13(self):
        self.print_header("Week 13: Electromagnetic Waves")
        
        # Problem 1
        f = 100e6
        wavelength = self.c / f
        print(f"1) Wavelength = {wavelength:.2f} m")

        # Problem 2
        E0 = 120
        I_avg = (E0**2) / (2 * self.mu_0 * self.c)
        print(f"2) Intensity I_avg = {I_avg:.2f} W/m^2")

        # Problem 3
        P = 15; diameter = 4.0e-3
        area = math.pi * (diameter / 2)**2
        I_laser = P / area
        # Assuming total absorption (pressure = I / c)
        pressure = I_laser / self.c
        force = pressure * area
        print(f"3) Intensity = {I_laser:.2e} W/m^2, Pressure = {pressure:.2e} Pa, Force = {force:.2e} N")

    def run_all(self):
        self.solve_week_1()
        self.solve_week_2()
        self.solve_week_3()
        self.solve_week_4()
        self.solve_week_5()
        self.solve_week_6()
        self.solve_week_8()
        self.solve_week_9()
        self.solve_week_10()
        self.solve_week_11()
        self.solve_week_12()
        self.solve_week_13()

if __name__ == "__main__":
    solver = PhysicsSolver()
    solver.run_all()
```

```python
import math

class PhysicsCalculator:
    def __init__(self):
        # Universal Physical Constants
        self.k = 8.987551e9            # Coulomb's constant (N m^2 / C^2)
        self.eps_0 = 8.854187e-12      # Vacuum permittivity (C^2 / N m^2)
        self.mu_0 = 4 * math.pi * 1e-7 # Vacuum permeability (T m / A)
        self.e_charge = 1.602e-19      # Elementary charge (C)
        self.m_e = 9.109e-31           # Electron mass (kg)
        self.c = 2.9979e8              # Speed of light (m/s)

    # ==========================================
    # Chapters 23-25: Electric Fields & Potential
    # ==========================================
    def point_charge_e_field(self, q, r):
        """Calculates the electric field magnitude of a point charge."""
        return self.k * abs(q) / (r**2)

    def electron_acceleration_uniform_field(self, delta_V):
        """Calculates final speed of an electron released from rest across a potential difference."""
        return math.sqrt(2 * self.e_charge * delta_V / self.m_e)

    def point_charge_flux(self, q):
        """Calculates electric flux through a closed surface enclosing charge q."""
        return q / self.eps_0

    def point_charge_potential(self, q, r):
        """Calculates the electric potential at distance r from charge q."""
        return self.k * q / r

    def uniform_surface_charge_field(self, sigma):
        """Calculates the electric field of an infinite sheet of charge."""
        return sigma / (2 * self.eps_0)

    # ==========================================
    # Chapter 26: Capacitance and Dielectrics
    # ==========================================
    def parallel_plate_capacitance(self, area, distance):
        """Calculates capacitance of a parallel plate capacitor."""
        return self.eps_0 * area / distance

    def capacitor_energy(self, C, V=None, Q=None):
        """Calculates energy stored in a capacitor given C and either V or Q."""
        if V is not None:
            return 0.5 * C * (V**2)
        elif Q is not None:
            return 0.5 * (Q**2) / C
        else:
            raise ValueError("Must provide either Voltage (V) or Charge (Q)")

    def insert_dielectric(self, C0, kappa, V_constant=True, V0=None, Q0=None):
        """Calculates new C, Q, V, and U when a dielectric is inserted."""
        C_new = kappa * C0
        if V_constant and V0 is not None:
            Q_new = C_new * V0
            U_new = 0.5 * C_new * (V0**2)
            return {"C": C_new, "Q": Q_new, "V": V0, "U": U_new}
        elif not V_constant and Q0 is not None:
            V_new = Q0 / C_new
            U_new = 0.5 * (Q0**2) / C_new
            return {"C": C_new, "Q": Q0, "V": V_new, "U": U_new}

    # ==========================================
    # Chapters 27-28: DC Circuits & Resistance
    # ==========================================
    def ohms_law(self, V=None, I=None, R=None):
        """Solves for the missing variable in Ohm's Law (V = IR)."""
        if V is None: return I * R
        if I is None: return V / R
        if R is None: return V / I

    def resistance_temperature_dependence(self, R0, alpha, delta_T):
        """Calculates resistance change due to temperature."""
        return R0 * (1 + alpha * delta_T)

    def equivalent_resistance(self, resistors, connection_type="series"):
        """Calculates equivalent resistance for a list of resistors."""
        if connection_type == "series":
            return sum(resistors)
        elif connection_type == "parallel":
            return 1 / sum(1/r for r in resistors)

    # ==========================================
    # Chapters 29-30: Magnetic Fields & Sources
    # ==========================================
    def lorentz_force_electron(self, v, B, theta_degrees=90):
        """Calculates magnetic force on a moving electron."""
        theta = math.radians(theta_degrees)
        force = self.e_charge * v * B * math.sin(theta)
        radius = (self.m_e * v) / (self.e_charge * B) if theta_degrees == 90 else None
        return {"Force": force, "Orbit Radius": radius}

    def b_field_long_wire(self, I, r):
        """Calculates magnetic field at distance r from a long straight wire."""
        return (self.mu_0 * I) / (2 * math.pi * r)

    def b_field_solenoid(self, turns_per_meter, I):
        """Calculates magnetic field inside an ideal solenoid."""
        return self.mu_0 * turns_per_meter * I

    # ==========================================
    # Chapters 31-32: Faraday's Law & Inductance
    # ==========================================
    def faradays_law(self, N, flux_initial, flux_final, dt):
        """Calculates average induced EMF."""
        dFlux = flux_final - flux_initial
        return -N * (dFlux / dt)

    def motional_emf(self, B, L, v):
        """Calculates EMF induced by a conductor moving through a B-field."""
        return B * L * v

    def rl_circuit_current(self, V, R, L, t):
        """Calculates current in an RL circuit at time t after connecting battery."""
        tau = L / R
        return (V / R) * (1 - math.exp(-t / tau))

    # ==========================================
    # Chapters 33-34: AC Circuits & EM Waves
    # ==========================================
    def rc_ac_impedance(self, R, C, f):
        """Calculates capacitive reactance and total impedance."""
        omega = 2 * math.pi * f
        Xc = 1 / (omega * C)
        Z = math.sqrt(R**2 + Xc**2)
        return {"Xc": Xc, "Z": Z}

    def rlc_resonance(self, L, C):
        """Calculates resonant frequency of an RLC circuit."""
        return 1 / (2 * math.pi * math.sqrt(L * C))

    def em_wave_properties(self, f=None, wavelength=None):
        """Converts between frequency and wavelength for EM waves."""
        if f is not None: return self.c / f
        if wavelength is not None: return self.c / wavelength

    def radiation_pressure(self, power, area, perfectly_reflecting=False):
        """Calculates intensity and radiation pressure."""
        intensity = power / area
        pressure = (2 * intensity / self.c) if perfectly_reflecting else (intensity / self.c)
        return {"Intensity": intensity, "Pressure": pressure, "Force": pressure * area}


# ==========================================
# EXAMPLE USAGE
# ==========================================
if __name__ == "__main__":
    calc = PhysicsCalculator()

    print("--- Testing General Functions with Week 1 & 4 Problems ---")
    
    # Week 1, Problem 1: E-field of point charge
    q = 3.0e-9  # 3.0 nC
    r = 0.20    # 0.20 m
    E = calc.point_charge_e_field(q, r)
    print(f"Week 1.1) E-field: {E:.2f} N/C")

    # Week 4, Problem 3: Dielectric Insertion (Connected to source)
    C0 = 150e-12  # 150 pF
    kappa = 4.2
    V_source = 200
    results_connected = calc.insert_dielectric(C0, kappa, V_constant=True, V0=V_source)
    print(f"Week 4.3a) Connected -> C: {results_connected['C']:.2e} F, Q: {results_connected['Q']:.2e} C")

    # Week 6, Problem 2: Parallel Resistors
    resistors = [3, 6, 9]
    R_eq = calc.equivalent_resistance(resistors, connection_type="parallel")
    print(f"Week 6.2) Equivalent Parallel Resistance: {R_eq:.2f} Ohms")

    # Week 13, Problem 1: EM Wave Equation
    f = 100e6 # 100 MHz
    wavelength = calc.em_wave_properties(f=f)
    print(f"Week 13.1) EM Wavelength: {wavelength:.2f} m")
```
